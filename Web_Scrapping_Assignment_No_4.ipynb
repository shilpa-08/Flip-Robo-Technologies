# Q no 1 

1. Scrape the details of most viewed videos on YouTube from Wikipedia:
    
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos     
        
You need to find following details: 
    
    A) Rank
    B) Name 
    C) Artist 
    D) Upload date 
    E) Views 

```python
# importing necessary libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
import time


# sending the request to the webpage server to get the source code of 
page=requests.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ')
time.sleep(2)


# getting the page content in the soup object
soup=BeautifulSoup(page.content)


# finding the table data into table variable
table= soup.find('table',{'class':"wikitable sortable"})


# taking all the rows from table
rows=table.find_all('tr')


row_list=[]                                      # declaring empty row_list to save the row data
for tr in rows:
    td=tr.find_all('td')                         # finding all the table data in a variable
    row_data=[i.text for i in td]
    row_list.append(row_data)                    # appending all the data in a list
    
    
# creating the dataframe
you_tube_videos=pd.DataFrame(row_list,columns=['Rank','Name','Artist','Views','Upload Date ',6])      
you_tube_videos.drop(6,axis=1,inplace=True)             # dropping an extra column of the table
you_tube_videos.replace('None',np.nan)                  # replacing 'None' values with np.nan
you_tube_videos.dropna()                                # dropping nan values
```

# Q NO 2

            B) Series
            C) Place 
            D) Date
            E) Time
            
            Note: - From bcci.tv home page you have reach to the international fixture page through code. 

```python
# importing necessary libraries
import pandas as pd
import selenium
from selenium import webdriver
import time



# setting up chrome driver
driver=webdriver.Chrome(r'Chromedriver.exe')
time.sleep(4)


# loading the website
driver.get('https://www.bcci.tv/.')
time.sleep(2)



# clicking on international tab
International_tab=driver.find_elements_by_xpath('//li[@data-nav-index="0"]')
for i in International_tab:
    if i.text =="INTERNATIONAL":
        i.click()
time.sleep(1)

        
        
# clicking on fixture option in the internatinal_tab list
fixture_button=driver.find_elements_by_xpath('//a[@href="/international/fixtures"]')
for i in  fixture_button:
    if i.text=="Fixtures":
        i.click()
time.sleep(2)        




# defining empty list for storing output
date=[]
day=[]
month=[]
time=[]
place=[]
match_title=[]
series=[]


# retrieving data using xpath
for i in driver.find_elements_by_xpath('//div[@class="fixture__datetime desktop-only"]'):
    temp_data=i.text.split('\n')
    date.append(temp_data[1]+" "+temp_data[2]+" "+temp_data[0])
    time.append(temp_data[3])

for i in driver.find_elements_by_xpath('//div[@class="fixture__info u-skewed"]'):
    temp_data=i.text.split('\n')
    match_title.append(temp_data[2])
    place.append(temp_data[3])
    series.append(temp_data[0]+" "+temp_data[1])

    
    
# creating DataFrame
International_fixtures=pd.DataFrame({})
International_fixtures['Match_title']=match_title
International_fixtures['Series']=series
International_fixtures['Place']=place
International_fixtures['Date']=date
International_fixtures['Time']=time


# calling the dataframe
International_fixtures
```

# Q NO 5

 Scrape the details of trending repositories on Github.com. Url  = https://github.com/
        
        You have to find the following details:
            A) Repository title 
            B) Repository description
            C) Contributors count
            D) Language used 
            
    
    Note: - From the home page you have to click on the trending option from Explore menu through code. 


```python
# importing necessary libraries

import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


# opening chrome 
driver=selenium.webdriver.Chrome(r'Chromedriver.exe')


# loading the website
driver.get('https://github.com/')


#  clicking on explore button
explore_button_link=driver.find_element_by_xpath('//a[@data-ga-click="(Logged out) Header, go to Explore"]')
explore_link=explore_button_link.get_attribute('href')
driver.get(explore_link)


# declaring empty list to store output
repository_title=[]
repository_link=[]
repository_description=[]
languages_used=[]
for i in driver.find_elements_by_xpath('//a[@class="text-bold"]'):
    repository_title.append(i.text)
    link=i.get_attribute('href')
    repository_link.append(link)
    


for i in repository_link:
    driver.get(i)
    try :
        description=driver.find_element_by_xpath('//p[@class="f4 mt-3"]')
        repository_description.append(description.text)
    except NoSuchElementException:
        repository_description.append(' -- ')
        


for i in repository_link:
    driver.get(i)
    try :
        languages=''
        for i in driver.find_elements_by_xpath('//li//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]'):
            languages=languages+(i.text.replace('\n',' ').replace('%','% , '))
            #print(languages)
            #print(i.text)
        languages_used.append(languages)
    except NoSuchElementException:
        languages_used.append(' -- ')
        
contributors_count=[]
contributors=''
for i in repository_link:
    driver.get(i)
    try :
        contributors=''
        for i in driver.find_elements_by_xpath('//a[@class="Link--primary no-underline"]'):
            contributors=contributors+(i.text.replace('\n',' '))
        contributors_count.append(contributors)
    except NoSuchElementException:
        contributors_count.append(' -- ')
        
count=[]
for i in contributors_count:
    contributors=i.split('Contributors')
    try :
        count.append(contributors[1])
    except IndexError:
        count.append('-')
    

    
# Creating dataframe
github=pd.DataFrame({})
github['Repository Title']=repository_title
github['Repository_description']=repository_description
github['Contributors_count']=count
github['Languages_used']=languages_used

github
```

```python
# closing the chrome driver
driver.close()
```

        Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code 
 

```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver


# opening chrome driver
driver=webdriver.Chrome(r'Chromedriver.exe')


# loading the website
driver.get('https://www.naukri.com/')


# clicking on recruiters_tab
recruiters_tab=driver.find_element_by_xpath('//a[@title="Search Recruiters"]')
link=recruiters_tab.get_attribute('href')
driver.get(link)



# inserting data in search tab
search_bar=driver.find_element_by_xpath('//input[@class="sugInp"]')
search_bar.send_keys('Data Science')


# clicking on search button
search_button=driver.find_element_by_id('qsbFormBtn')
search_button.click()


# scrapping the name of the recruiters
name=[]
for i in driver.find_elements_by_xpath('//span[@class="fl ellipsis"]'):
    name.append(i.text)
    
# scrapping the designation
designation=[]
for i in driver.find_elements_by_xpath('//span[@class="ellipsis clr"]'):
    designation.append(i.text)
    
# scrapping the company names
whole_data=[]        # as there is no unoque class for company so taking the whole data as a list and splitting it
for i in driver.find_elements_by_xpath('//p[@class="highlightable"]'):
    whole_data.append(i.text)
company_name=[]
location=[]
for i in whole_data:
    data=i.split('\n')
    try:
        company=data[2]
        company_name.append(company)
        place=data[3]
        location.append(place)
    except IndexError:
        location.append('--')
    
    
# scrapping the skills
skills=[]
for i in driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]'):
    skills.append(i.text)
    
    
# creating dataframe
data_science_recruiters=pd.DataFrame({})
data_science_recruiters['Name']=name
data_science_recruiters['Designation']=designation
data_science_recruiters['Company']=company_name
data_science_recruiters['Skills they hire for']=skills
data_science_recruiters['Location']=location


data_science_recruiters
```

# Q no 9

. Scrape the details most watched tv series of all time from imdb.com.   Url =  https://www.imdb.com/list/ls095964455/ 
    
    You have to find the following details: A) Name B) Year span C) Genre D) Run time E) Ratings F) Votes 
 

```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
import requests
from bs4 import BeautifulSoup



# initaiting chrome browser
driver=webdriver.Chrome(r'Chromedriver.exe')
time.sleep(3)


# loading the website
driver.get('https://www.imdb.com/list/ls095964455/')
time.sleep(2)



# getting page content using beautifulsoup
page=requests.get('https://www.imdb.com/list/ls095964455/')
soup=BeautifulSoup(page.content)
data=soup.find_all('div',class_="lister-item-content")            # getting all the output in data variable
# declaring empty variables
name=[]
year=[]
genre=[]
runtime=[]
ratings=[]
votes=[]
for i in data:
    temp_name=i.find('a')                 # searching for name
    name.append(temp_name.text)
    temp_year=i.find('span',class_="lister-item-year text-muted unbold")               # searching for year
    year.append(temp_year.text)
    temp_genre=i.find('span',class_="genre")          # searching for genre
    genre.append(temp_genre.text.replace('\n','').strip())
    temp_time=i.find('span',class_="runtime")                   ## searching for time
    runtime.append(temp_time.text)
    temp_rating=i.find('span',class_="ipl-rating-star__rating")       ## searching for ratings
    ratings.append(temp_rating.text)
    #temp_vote=i.find('span',name_="nv")
    #votes.append(temp_vote.text)

    

for i in driver.find_elements_by_xpath('//span[@name="nv"]'):       # # searching for votes
    votes.append(i.text)
    

# creating dataframe

series=pd.DataFrame({})
series['Name']=name
series['Year Span']=year
series['Genre']=genre
series['Time']=time
series['Run Time']=runtime
series['Ratings']=ratings
series['Votes']=votes

series
```
