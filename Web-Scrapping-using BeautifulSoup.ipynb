# Assignment Question No 1:

##### Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’. 

```python
#importing necessary libraries
from bs4 import BeautifulSoup
import requests
```

```python
# requesting on page 
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
# reading content of page
#page.content
```

```python
soup=BeautifulSoup(page.content)
# print(soup)
```

```python
# beautifying content in proper look
#print(soup.prettify())
```

```python
#First we will extract the HTML tag where we have the first job title 
header_h1 =soup.find_all('h1',class_='firstHeading')
header_h1
```

```python
# converting header_h1 into list
Header_H1=[]
for i in header_h1:
    Header_H1.append(i.text)
Header_H1
```

```python
#extracting second header column
header_h2 =soup.find_all('h2',class_='mp-h2')
header_h2
```

```python
# converting header_h2 into list
Header_H2=[]
for i in header_h2:
    Header_H2.append(i.text)
Header_H2
```

```python
#extracting third header column
header_h3 =soup.find_all('h3',class_='vector-menu-heading')
header_h3
```

```python
# converting header_h3 into list
Header_H3=[]
for i in header_h3:
    Header_H3.append(i.text.replace('\n',''))
Header_H3
```

```python
#extracting headlines
headline=soup.find_all('span',class_='mw-headline')
headline
```

```python
# converting headlines into list
Headlines=[]
for i in headline:
    Headlines.append(i.text)
Headlines
```

```python
page7=requests.get('https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees')
page7
```

```python
#page7.content
```

```python
soup7=BeautifulSoup(page7.content)
```

```python
#print(soup7.prettify())
```

```python
name2=soup7.find_all('h2',class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
name2
```

```python
p_name2=[]
for i in name2:
    p_name2.append(i.text.replace('\n',''))
p_name2
```

```python
mobile_name=[]
fname=''  
str2=')'
# declaring empty string for converting list item into individual list    
for i in range(0,len(p_name2)):
    names=p_name2[i]    # declaring individual item of list in a string
    if '|' in names:
        fname=names.split(')')
        mobile_name.append(fname[0]+str2)
    elif ')' in names:
        fname=names.split(')')
        mobile_name.append(fname[0]+str2)
    else:
        fname=names          # splitting the string into list
        mobile_name.append(fname+str2)
```

```python
mobile_name
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
  
htmldata = urlopen('https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees')
soup = BeautifulSoup(htmldata, 'html.parser')
images = soup.find_all('img')
```

```python
image_url=[]  
for item in images:
    #print(item['src'])
    image_url.append(item['src'])
```

```python
image_url
```

```python
#as the first three url are of some another images, not related to our answer, so deleting them
del image_url [0:3]
```

```python
image_url
```

Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.  
 

```python
page8=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YJL0dLUzZPY')
page8
```

```python
#page8.content
```

```python
soup8=BeautifulSoup(page8.content)
```

```python
#print(soup8.prettify())
```

```python
period=soup8.find_all('p',class_='period-name')
period
```

```python
period_name=[]
for i in period:
    period_name.append(i.text)
period_name
```

```python
short_Descr=soup8.find_all('p',class_='short-desc')
short_Descr
```

```python
short_description=[]
for i in short_Descr:
    short_description.append(i.text)
short_description
```

```python
temp=soup8.find_all('p',class_='temp')
temp
```

```python
temperature=[]
for i in temp:
    temperature.append(i.text)
temperature
```

```python
descr=soup8.find_all('div',class_='col-sm-10 forecast-text')
descr
```

```python
description=[]
for i in descr:
    description.append(i.text)
description
```

```python
del description[-4:]
```

```python
print(len(period_name),len(short_description),len(temperature),len(description))
```

```python
import pandas as pd
San_Francisco_weather_info=pd.DataFrame({})
San_Francisco_weather_info['Period']=period_name
San_Francisco_weather_info['Short_Description']=short_description
San_Francisco_weather_info['Temperature']=temperature
San_Francisco_weather_info['Description']=description

```
