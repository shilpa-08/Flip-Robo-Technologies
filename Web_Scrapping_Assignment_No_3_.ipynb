```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


# opening the chromedriver
driver=webdriver.Chrome(r'Chromedriver.exe')

# loading the website
driver.get('https://www.amazon.in/')
time.sleep(2)

# taking the input from the user
input_from_user=input("Enter the name of product you want to search")
```

```python
# entering the user's information in the search_box
search_bar=driver.find_element_by_id('twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys(input_from_user)
                     
# clicking the search button
search_button=driver.find_element_by_id('nav-search-submit-button')
search_button.click()

                     
```

```python
#  searching for the product_names and product_url on first three pages
product_name=[]
product_url=[]
for i in range(0,3):
    for i in driver.find_elements_by_xpath('//span[@class="a-size-base-plus a-color-base a-text-normal"]'):
        product_name.append(i.text)
    for i in driver.find_elements_by_xpath('//h2//a[@class="a-link-normal a-text-normal"]'):
        product_url.append(i.get_attribute('href'))
    try:
        driver.find_element_by_xpath('//li[@class="a-last"]').click()
        time.sleep(2)
    except NoSuchElementException :
        return
```

```python
#declaring empty lists for saving the information

No_of_Ratings=[]
Product_Price=[]
Availability=[]   
other_details=[]
brand_name=[]
expected_delivery=[]
Ratings=[]
                     


for i in product_url:
    driver.get(i)
    time.sleep(2)                    
    try:                                    # searching for number of ratings on each product_url
        no_rating=driver.find_element_by_xpath('//span[@id="acrCustomerReviewText"]')
        No_of_Ratings.append(no_rating.text)
    except NoSuchElementException:
        No_of_Ratings.append('-')

    try:                                    # searching for price on each product_url
        price=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-price priceBlockBuyingPriceString"]')
        Product_Price.append(price.text)
    except NoSuchElementException:
        Product_Price.append('-')
                    
    try:                                    # searching for availability on each product_url
        available=driver.find_element_by_xpath('//div[@id="availability"]')
        Availability.append(available.text)
    except NoSuchElementException:
        Availability.append('-')
                     
    try:                                       # searching for details on each product_url
        details=driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')
        other_details.append(details.text.replace('\n',','))
    except NoSuchElementException:
        other_details.append('-')

    try:                                      # searching for brand_name of ratings on each product_url
        name=driver.find_element_by_xpath('//tr[@class="a-spacing-small"]//td[2]')
        brand_name.append(name.text)
    except NoSuchElementException:
        brand_name.append('-')

    try:                              # searching for expected_delivery on each product_url
        delivered=driver.find_element_by_xpath('//span[@class="a-color-error"]')
        expected_delivery.append(delivered.text)
    except NoSuchElementException:
        expected_delivery.append('-')
        
    try:                             # searching for ratings on each product_url
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")#locating the ratingd link
        rate.click()                                                      #click the rating link found
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")#locating the rating
        Ratings.append(rating.text)#appending the ratings in Ratings list
    except NoSuchElementException   as e:
        Ratings.append("-")#appending the No rating if no rating is there

```

```python
# Creating dataframe

details=pd.DataFrame({})
details['Brand_name']=brand_name
details['Name_of_the_product']=product_name
details['Ratings']=Ratings
details['No_of_Ratings']=No_of_Ratings
details['expected_delivery']=expected_delivery
details['Availability']=Availability 
details['other_details']=other_details
details['product_url']=product_url
details['Product_Price']=Product_Price


```

```python
# saving the dataFrame into csv
details.to_csv('product_details.csv')
```

# Assignment Question No 4

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


# openinig the chromedriver
driver=webdriver.Chrome(r'Chromedriver.exe')


#Loading the website
driver.get('https://www.flipkart.com/')

# taking input as pixel4
input_from_user='redmi mobiles'


# sending the input in the search_box
search_box=driver.find_element_by_xpath("//input[@class='_3704LK']")
search_box.clear()
search_box.send_keys(input_from_user)
time.sleep(5)

# clicking the search_button
search_button=driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()
time.sleep(1)


# declaring empty lists for saving data
product_name=[]
product_url=[]
display=[]
Battery=[]
color=[]
primary_camera=[]
secondary_camera=[]
camera=[]
processor=[]
price=[]
Ram=[]
Rom=[]
Ram_Rom=[]

for i in driver.find_elements_by_xpath('//div[@class="_4rR01T"]'):
    product_name.append(i.text)                                        # appending product_name
    
for i in driver.find_elements_by_xpath('//a[@class="_1fQZEK"]'):
    product_url.append(i.get_attribute('href'))                         # appending product_url

    
# searching for the display information from product_urls
for i in product_url:
    driver.get(i)
    time.sleep(1)
    try:
        size=driver.find_element_by_xpath('//div[@class="_2418kt"]//ul//li[2]')
        display.append(size.text)
    except NoSuchElementException:
        display.append('-')

        
# searching for the battery information from product_urls
for i in product_url:
    driver.get(i)
    time.sleep(1)
    try:
        size=driver.find_element_by_xpath('//div[@class="_2418kt"]//ul//li[4]')
        Battery.append(size.text)
    except NoSuchElementException :
        Battery.append('-')
        
    try:                             # searching for the colors information from product_urls
        name=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        pname=(name.text.split('('))
        colors=pname[1].split(',')
        color.append(colors[0])
    except NoSuchElementException :
        product_name.append('-')

    try:                                 # searching for the camera's information from product_urls
        size=driver.find_element_by_xpath('//div[@class="_2418kt"]//ul//li[3]')
        camera=size.text.split('|')
        primary_camera.append(camera[0])
        secondary_camera.append(camera[1])
    except IndexError or NoSuchElementException :
        secondary_camera.append('-')
      
    try:                          #   searching for the processor information from product_urls
        size=driver.find_element_by_xpath('//div[@class="_2418kt"]//ul//li[5]')
        processor.append(size.text)
    except NoSuchElementException:
        processor.append('-')
        
    try:                            # searching for the price information from product_urls
        rs=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(rs.text)
    except NoSuchElementException:
        price.append('-')

    try:                               # searching for the RAM &ROM information from product_urls
        memory=driver.find_element_by_xpath('//div[@class="_2418kt"]//ul//li[1]')
        Ram_Rom=memory.text.split('|')
        Ram.append(Ram_Rom[0])
        Rom.append(Ram_Rom[1])
    except NoSuchElementException:
        Ram.append('-')
        Rom.append('-')

```

```python
# Creating The DataFrame

smartphone_details=pd.DataFrame({})
smartphone_details['Smartphone_name']=product_name
smartphone_details['Color']=color
smartphone_details['RAM']=Ram
smartphone_details['Storage (ROM)']=Rom
smartphone_details['Primary_Camera']=primary_camera
smartphone_details['Secondary=_Camera']=secondary_camera
smartphone_details['Product_URL']=product_url
smartphone_details['Display']=display
smartphone_details['Price']=price
smartphone_details['Battery']=Battery
smartphone_details['Processor']=processor
```

```python
# saving the DataFrame into the csv file
smartphone_details.to_csv('smartphone_details.csv')
driver.close()
```

5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps

```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


# opening the chromedriver
driver=webdriver.Chrome(r'Chromedriver.exe')


# opening the google maps
driver.get('https://www.google.com/maps/@20.5446791,74.5345989,15z')
time.sleep(2)


# getting input from the user
input_from_user=input("Enter the name of product you want to search = ")
```

```python

# sending the input in the search_box
search_bar=driver.find_element_by_xpath('//input[@autofocus="autofocus"]')
search_bar.send_keys(input_from_user)



# clicking on the search button
search_button=driver.find_element_by_xpath('//button[@class="searchbox-searchbutton"]')
search_button.click()



# getting the current url as longitude and latitude of the places are mentioned in the url itself
url=driver.current_url
print("URL = ",url)


# splitting url with '@'
add=url.split('@')


# again splitting url with ','
address=add[1].split(',')


#getting the longitude and latitude
latitude=address[0]
longitude=address[1]

print('Latitude = ',latitude)
print('Longitude = ',longitude)
        

# closing the chrome window
driver.close()
```

# Assignment Question No 7

 Write a program to scrap all the available details of top 10 gaming laptops from digit.in

```python
# importing necessary libraries
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


# opening the website 
driver=webdriver.Chrome(r'Chromedriver.exe')

# opening the website 
driver.get('https://www.digit.in/')
driver.maximize_window()                       # maximizing the webpage 
time.sleep(2)


# clicking on the laptop tab 
laptop_tab=driver.find_element_by_xpath('//div[@class="menu"]//li[3]')
laptop_tab.click()


# clicking on the gaming laptop on side bar
gaming_laptop_button=driver.find_element_by_xpath('//ul[@class="list-unstyled sidebar-list"]//li[10]')
gaming_laptop_button.click()


# clicking on the view_all_button
view_all_button=driver.find_element_by_xpath('//a[@href="https://www.digit.in/hotdeal/laptops-deals/"]')
view_all_button.click()
time.sleep(2)


# defining empty lists for storing information
product_name=[]
product_image=[]
price=[]

# searching elements through xpath
for i in driver.find_elements_by_xpath('//div[@class="hotdeal-block"]'):
    product_name.append(i.text)
time.sleep(2)
for i in driver.find_elements_by_xpath('//div[@class="hotdeal-block"]//span//img'):
    img=i.get_attribute('src')
    product_image.append(img)
time.sleep(2)
for i in driver.find_elements_by_xpath('//div[@class="deal-price"]'):
    price.append(i.text)
    
    
# creating the DataFrame
Gaming_laptops=pd.DataFrame({})
Gaming_laptops['Product_Name']=product_name
Gaming_laptops['Product_Image']=product_image
Gaming_laptops['Product_Price']=price



#closing the chrome driver
driver.close()
```
